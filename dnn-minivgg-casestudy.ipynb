```python
import tensorflow as tf
from tensorflow.keras import layers, models
import numpy as np
import matplotlib.pyplot as plt
import os
from sklearn.model_selection import train_test_split
import random
import warnings

warnings.filterwarnings("ignore")

# -------------------------
# 1. Load all images and labels manually (SAFE VERSION)
# -------------------------

root_path = "/kaggle/input/asl-dataset/asl_dataset"
IMG_SIZE = 32

images = []
labels = []
class_names = sorted(os.listdir(root_path))

print("Detected Classes:", class_names)

valid_ext = (".png", ".jpg", ".jpeg", ".bmp")

for class_index, class_name in enumerate(class_names):
    class_folder = os.path.join(root_path, class_name)

    # Skip empty directories
    if len(os.listdir(class_folder)) == 0:
        print(f"⚠️ Skipping empty folder: {class_name}")
        continue
    
    for img_name in os.listdir(class_folder):
        img_path = os.path.join(class_folder, img_name)

        if os.path.isdir(img_path):
            continue
        if not img_name.lower().endswith(valid_ext):
            continue

        try:
            img = tf.keras.preprocessing.image.load_img(
                img_path,
                target_size=(IMG_SIZE, IMG_SIZE),
                color_mode="grayscale"
            )
            img_arr = tf.keras.preprocessing.image.img_to_array(img)
            images.append(img_arr)
            labels.append(class_index)

        except:
            print(f"Skipping corrupted file: {img_path}")
            continue


images = np.array(images) / 255.0
labels = tf.keras.utils.to_categorical(labels, num_classes=len(class_names))

print("Total Images Loaded:", images.shape)

# -------------------------
# 2. Manual Train/Test Split
# -------------------------

X_train, X_test, y_train, y_test = train_test_split(
    images, labels, test_size=0.2, random_state=42, shuffle=True
)

print("Train:", X_train.shape, " Test:", X_test.shape)

# -------------------------
# 3. Build Mini-VGG Model (Advanced Architecture)
# -------------------------

model = models.Sequential([

    # --- Block 1 ---
    layers.Conv2D(32, (3,3), padding="same", activation='relu', input_shape=(IMG_SIZE, IMG_SIZE, 1)),
    layers.Conv2D(32, (3,3), padding="same", activation='relu'),
    layers.MaxPooling2D((2,2)),

    # --- Block 2 ---
    layers.Conv2D(64, (3,3), padding="same", activation='relu'),
    layers.Conv2D(64, (3,3), padding="same", activation='relu'),
    layers.MaxPooling2D((2,2)),

    # --- Block 3 ---
    layers.Conv2D(128, (3,3), padding="same", activation='relu'),
    layers.Conv2D(128, (3,3), padding="same", activation='relu'),
    layers.MaxPooling2D((2,2)),

    # --- Classification Head ---
    layers.Flatten(),
    layers.Dense(256, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(len(class_names), activation='softmax')
])

model.summary()

# -------------------------
# 4. Compile Model
# -------------------------

model.compile(
    optimizer=tf.keras.optimizers.Adam(0.0005),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

# -------------------------
# 5. Train Model
# -------------------------

history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=15,
    batch_size=32
)

# -------------------------
# 6. Evaluate Model
# -------------------------

loss, acc = model.evaluate(X_test, y_test)
print(f"\nTest Accuracy: {acc*100:.2f}%")

# -------------------------
# 7. Plot Accuracy and Loss
# -------------------------

plt.figure(figsize=(14,5))

plt.subplot(1,2,1)
plt.plot(history.history['accuracy'], label='Train Acc')
plt.plot(history.history['val_accuracy'], label='Test Acc')
plt.title("Accuracy")
plt.legend()

plt.subplot(1,2,2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Test Loss')
plt.title("Loss")
plt.legend()

plt.show()

# -------------------------
# 8. Predict on 2–3 Random Test Images
# -------------------------

print("\n🔍 Predictions on Random Test Images:\n")

plt.figure(figsize=(10,4))

rand_indices = random.sample(range(len(X_test)), 3)

for i, idx in enumerate(rand_indices):
    img = X_test[idx]
    true_label = class_names[np.argmax(y_test[idx])]

    pred = model.predict(np.expand_dims(img, axis=0), verbose=0)
    pred_label = class_names[np.argmax(pred)]

    print(f"Image {i+1}:  True = {true_label} | Pred = {pred_label}")

    plt.subplot(1, 3, i+1)
    plt.imshow(img.squeeze(), cmap='gray')
    plt.title(f"True: {true_label}\nPred: {pred_label}")
    plt.axis("off")

plt.tight_layout()
plt.show()

# -------------------------
# 9. Predict Custom Image
# -------------------------

def predict_image(path):
    img = tf.keras.preprocessing.image.load_img(
        path, 
        target_size=(IMG_SIZE, IMG_SIZE), 
        color_mode="grayscale"
    )
    img_arr = tf.keras.preprocessing.image.img_to_array(img) / 255.0
    img_arr = np.expand_dims(img_arr, axis=0)

    pred = model.predict(img_arr, verbose=0)
    class_index = np.argmax(pred)

    print("Predicted Class:", class_names[class_index])
    return class_index

# Custom example:
predict_image("/kaggle/input/asl-dataset/asl_dataset/1/hand1_1_bot_seg_1_cropped.jpeg")
```
